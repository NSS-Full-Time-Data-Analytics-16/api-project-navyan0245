```python
import requests
import pandas as pd
import matplotlib.pyplot as plt
```

1. Use the API to get all available data for the _GDP per capita, PPP (constant 2017 international $)_ indicator. Hint: this indicator has code "NY.GDP.PCAP.PP.KD". Adjust the query parameters so that you can retrieve all available rows. Convert the results to a DataFrame.

```python
endpoint = 'https://api.worldbank.org/v2/country/all/indicator/NY.GDP.PCAP.PP.KD'
params = {'format':'json','per_page':17290}
response = requests.get(endpoint,params=params)
```

```python
response
```

```python
res = response.json()
```

```python
gdp_data =pd.json_normalize(res[1])
gdp_data
```

```python
type(res)
```

```python
res[0].keys()
```

```python
pages = res[0]['pages']
pages
```

```python
gdp_df=pd.DataFrame(gdp_data)
gdp_df
```

```python
gdp_df.columns
```

```python
gdp_df=gdp_df.drop(columns=['unit','obs_status','decimal','indicator.id','indicator.value','countryiso3code'])
```

```python
gdp_df = gdp_df.rename(columns ={'country.id':'country_id','date':'year','value':'gdp_per_capita','country.value':'country_name'})
```

```python
gdp_df
```

2. Now, use the API to get all available data for _Life expectancy at birth, total (years)_. This indicator has code "SP.DYN.LE00.IN". Again, convert the results to a DataFrame.


```python
endpoint = 'https://api.worldbank.org/v2/country/all/indicator/SP.DYN.LE00.IN'
params = {'format':'json','per_page':17290}
response = requests.get(endpoint,params=params)
```

```python
res = response.json()
```

```python
lifexp_data =pd.json_normalize(res[1])
lifexp_data
```

```python
type(res)
```

```python
res[0].keys()
```

```python
lifexp_df=pd.DataFrame(lifexp_data)
lifexp_df
```

```python
lifexp_df=lifexp_df.drop(columns={'unit','obs_status','decimal','indicator.id','indicator.value','countryiso3code'})
```


3. Merge the two results DataFrames together. You may want to rename or drop columns prior to merging.

```python
gdp_life_exp = pd.merge(gdp_df,lifexp_df,on = ['country_id','year','country_name'])
```

```python
gdp_life_exp
```

#4. You can also get more information about the available countries (region, capital city, income level classification, etc.) by using the [Country API](https://datahelpdesk.worldbank.org/knowledgebase/articles/898590-country-api-queries). Use this API to pull in all available data. Merge this with your other datasets. Use this to now remove the rows that correspond to regions and not countries.

```python
endpoint = 'https://api.worldbank.org/v2/country'
params = {'format':'json','per_page':296}
response = requests.get(endpoint,params=params)
```

```python
res = response.json()
```

```python
res[0].keys()
```

```python
country_data =pd.json_normalize(res[1])
country_data
```

```python
country_df=pd.DataFrame(country_data)
country_df
```

```python
country_df.columns
```

```python
country_df=country_df.rename(columns={'name':'country_name','iso2Code':'country_id','incomeLevel.value':'incomeLevel_value'})
```

```python
country_df=country_df.drop(columns={'adminregion.id','id','adminregion.iso2code','adminregion.value','incomeLevel.iso2code','lendingType.id','lendingType.iso2code','region.id','region.iso2code','region.value','incomeLevel.id','lendingType.value'})
```

```python
country_df
```

```python
country_gdp_lifexp = pd.merge(gdp_life_exp,country_df,on=['country_id','country_name'])
```

```python
country_gdp_lifexp
```
